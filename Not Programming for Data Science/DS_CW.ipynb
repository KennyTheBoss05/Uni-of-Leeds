```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("THE_ULTIMATE_CLEAN_DATASET.csv")
survey=pd.read_csv("THE_ULTIMATE_CLEAN_DATASET_GoodOrBad.csv") 
```

```python
# Number of columns
num_columns = df.shape[1]
print("Number of columns:", num_columns)

# Number of rows
num_rows = df.shape[0]
print("Number of rows:", num_rows)

# Example values
print("Example values:")
print(df.head())

```

```python
plt.figure(figsize=(12, 8))


sns.histplot(df['A2'], bins=20, kde=True, color='skyblue', edgecolor='black')


plt.title('Age Distribution of Players', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(range(0, 101, 5), [str(i) for i in range(0, 101, 5)])

plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
sns.set(style='whitegrid')

# Box plot 
plt.figure(figsize=(12, 8))
sns.boxplot(x='A2', y='Weighted Score', data=survey)
plt.title('Environmental Perception by Biological Sex')
plt.xlabel('Biological Sex')
plt.ylabel('Agreement Level')
plt.show()
```

 *The median environmental perception score for females is around 4.0.
 
 *The estimated median environmental perception score for males is around 3.5.

Females tend to provide slightly higher environmental perception scores compared to males.
